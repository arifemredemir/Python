**1-) center()**
  - center() method returns the centered string using the specified position. 
  - usage : string.center(space(int),"specified character")
  

```python
text = "Hello Dear Welcome to My Gitub"
print(text.center(50,"-"))
---------------------------------------------------------------------------------------------------------------------------
Output: ----------Hello Dear Welcome to My Gitub----------

```

**2-) count()**
- count() method returns the number of times value is in string

```python
text = "Hello Dear Welcome to My Gitub"
print(text.count('o'))
---------------------------------------------------------------------------------------------------------------------------
Output: 3
```

**3-) find()**
- find() method returns the index of specified value 
- if it doesn't find specified value, it returns -1

```python
text = "Hello Dear Welcome to My Gitub"
print(text.find('o'))
---------------------------------------------------------------------------------------------------------------------------
Output: 4
```

**4-) capitalize()**
- capitalize() method converts the first character to upper case and other charachters to lower case

```python
text = "Hello Dear Welcome to My Gitub"
print(text.capitalize())
---------------------------------------------------------------------------------------------------------------------------
Output: Hello dear welcome to my repo
```

**5-) casefold()**
- casefold() methods converts the all characters to lower case

```python
text="HeLLo DeAR"
print(text.casefold())
---------------------------------------------------------------------------------------------------------------------------
Output: hello dear
```
**6-) title()**
- title() method converts first charachter of each word to upper case.

 ```python
text="may the force Be with you"
print(text.title())
---------------------------------------------------------------------------------------------------------------------------
Output: May The Force Be With You
```
**7-) upper()**
- upper() method converts the all string to upper case

```python
text="hello how are you today?"
print(text.upper())
---------------------------------------------------------------------------------------------------------------------------
Output:HELLO HOW ARE YOU TODAY?

```
**8-) lower()**
- lower() method converts the all character of string to lower case. 
- it is the same with casefold() method.

```python
content="HELLO HOW ARE YOU TODAY?"
print(content.lower())
---------------------------------------------------------------------------------------------------------------------------
Output: hello how are you today?

```
**9-) swapcase()**
- swapcase() method converts the upper case characters to lower case and the lower case characters to upper case.

```python
content="Hello How are YOU Today?"
print(content.swapcase())
---------------------------------------------------------------------------------------------------------------------------
Output: hELLO hOW ARE you tODAY?
```

**11-) index()**
- index() method returns the index of specified value
- if there are more than one, it returns the index of first character.
- it is almost the same with the find() method but the different:
- if The index method isn't found in string, it raises exception but if the find() method isn't found in string, it returns -1

```python
text="Hello World"
print(text.index("o"))
---------------------------------------------------------------------------------------------------------------------------
Output: 4 
```
**12-) startswith()**
- startswith() method returns true or false according to if the string starts with specified value
- you can specify more than one character.

```python
text="I like to read the article"
print(text.startswith('I'))
---------------------------------------------------------------------------------------------------------------------------
Output: True
```

**13-) endswith()**
- endswith() method returns true or false according to if the string ends with specified value
- you can specify more than one character.
```python
text="I like to read the article"
print(text.endswith('le'))
---------------------------------------------------------------------------------------------------------------------------
Output: True
```

**14-) expandtabs()**
- expantabs() method sets tab size of string to specified value.
- default is 0 

```python
text="H\tT\tM\tL"
print(text.expandtabs(5))
---------------------------------------------------------------------------------------------------------------------------
Output: H    T    M    L
```
**15-) isalnum()**
- the isalnum() method returns true if all characters of string are alphanumeric

```python
content="arifemre29"
state=content.isalnum()
print(state)
---------------------------------------------------------------------------------------------------------------------------
Output: True
```
**16-) isalpha()**
- isalpha() method returns true if all string characters are in alphabet

```python
content="arifemre29"
state=content.isalpha()
print(state)
---------------------------------------------------------------------------------------------------------------------------
Output: False
```

**17-) join()**
- join() methods join the specified value into end of the iterable string

```python
content = ("the girls\t","\tthe mans")
content[0].expandtabs(3)
content[1].expandtabs(3)
new_content=">".join(content)
print(new_content)
---------------------------------------------------------------------------------------------------------------------------
Output: the girls	>	the mans
```
**18-) split()**
- split() method splits string than specified value and returns it in list. 
- it takes 2 optional params, 1st is you will split it from which character and 2nd is how many split you.
- Usage: string.split('optional value','how many split you want')
- if you don't specify that how many split, it splits all of the string.

```python
text="text.hello.py.doc.xml"
text.split('.',3)
---------------------------------------------------------------------------------------------------------------------------
Output: ['text', 'hello', 'py', 'doc.xml']
```

**19-) rsplit()**
-rsplit() method is the same with split() method bur just starts spliting from right side.
```python
text="text.hello.py.doc.xml"
text.rsplit('.',3)
---------------------------------------------------------------------------------------------------------------------------
Output: ['text.hello', 'py', 'doc', 'xml']
```

**20-) splitlines()**
- splitlines() method splits the string at linebreaks and return it in list.
- the parameter value determines whether to include line break after split.
- Usage: string.splitlines(True/False)
- True case: to include linebreak at the ends of first line
- False case: to not include linebreak

```python
content = "Hello how are you today\nI'm doing well thank you"
x=content.splitlines(False)
print(x)
---------------------------------------------------------------------------------------------------------------------------
Output: ['Hello how are you today', "I'm doing well thank you"]
```
**21-) strip()**
- the strip() method is one of the most common to use methods
- the strip() method removes leading and trailing characters
- if you want to specify value that you want to remove but if you don't specify any value, it removes all space at the begining and end of the string

1st Example:
```python
text = "      mercedes         "
text_1="the"
text_2="car"
print(text_1+text+text_2)
print(text_1+text.strip()+text_2)
---------------------------------------------------------------------------------------------------------------------------
Output: the      mercedes         car
        themercedescar
```
2nd Example:
```python
txt = ",,,,,rrttgg.....banana....rrr"

x = txt.strip(",.grt")

print(x)
---------------------------------------------------------------------------------------------------------------------------
Output: banana
```

**22-) rstrip()**
- rstrip() method is the same with strip() method but it just strips the right side

```python
txt = ",,,,,rrttgg.....banana....rrr"
x=txt.rstrip('.r')
print(x)
---------------------------------------------------------------------------------------------------------------------------
Output: ,,,,,rrttgg.....banana
```

**23-) replace()**
- replace() method replaces the specified value with other specified value. 
- Usage: string.replace('remove word','add word')
- replace method() removes the remove word into string and adds the add word into the string.
```python
text="Hello how are you doing"
print(text)
new_text=text.replace('doing','going')
print(new_text)
---------------------------------------------------------------------------------------------------------------------------
Output: Hello how are you doing
        Hello how are you going
```
**24-) partition()**
- partition() method parts the string and returns in tuple
- specified value returns in middle of tuple
- entire(whole) string is stored in first part of tuple.
```python
text="Do you think HTML is a programming language?"
parted_text=text.partition('HTML')
print(parted_text)
---------------------------------------------------------------------------------------------------------------------------
Output:('Do you think ', 'HTML', ' is a programming language?')
```
**25-) rpartition()**
- rpartition() method is the same with partition() method but it just starts parting from right side

```python
text="one for you and one for me and two for him?"
partedd_text=text.rpartition('for')
parted_text=text.partition('for')

print(f'rpartition: {partedd_text}')
print(f'partition: {parted_text}')
---------------------------------------------------------------------------------------------------------------------------
Output: rpartition: ('one for you and one for me and two ', 'for', ' him?')
        partition: ('one ', 'for', ' you and one for me and two for him?')
```
